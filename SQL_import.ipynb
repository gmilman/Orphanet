```python
#import the data to mySQL

def convert_pd_df_tosql(file_name,table_name,schema='orphanet'):
    import config
    pw=config.pw
    connection_string = 'mysql+pymysql://root:' + pw + '@127.0.0.1:3306/'
    engine = create_engine(connection_string)
    df = pd.read_csv(file_name, index_col=False)
    df.to_sql(table_name,engine,schema,index=False, chunksize=5000,if_exists='replace')
    return 'Created table'
```

```python
# Data cleaning - fixing incoherences between main disease table and clinical annotations table

def remove_code_anomalies(table_1,table_2, id_1,id_2):
    
    connection_string = 'mysql+pymysql://root:' + pw + '@127.0.0.1:3306/orphanet'
    engine = create_engine(connection_string)

    # load the clinical_annotations table and orpha disease table into a dataframe
    query = f'SELECT * FROM {table_1}'
    df_1 = pd.read_sql_query(query, engine)

    query = f'SELECT * FROM {table_2}'
    df_2 = pd.read_sql_query(query, engine)

    # get the OrphaCodes that are missing from the orpha_diseases table
    missing_codes = set(df_1[id_1]) - set(df_2[id_2])
    
    # drop the rows from the clinical_annotations table that have missing OrphaCodes
    df_1 = df_1[~df_1[id_1].isin(missing_codes)]
    
    #to check that codes removed
    missing_codes = set(df_1[id_1]) - set(df_2[id_2])

    #update the clinical_annotations table in the database
    df_1.to_sql(table_1, engine, if_exists='replace', index=False)
    
    return df_1[id_1].dtypes, df_2[id_2].dtypes, missing_codes
    
remove_code_anomalies('orpha_definitions', 'orpha_diseases', 'OrphaCode','OrphaCode')
```

```python
#Function to define froeign key

def define_FK(table,col_name_fk,  table_pk,col_name_pk):
    import config 
    pw=config.pw

    connection_string = 'mysql+pymysql://root:' + pw + '@127.0.0.1:3306/orphanet'
    engine = create_engine(connection_string)
    with engine.connect() as con:
        con.execute(f'ALTER TABLE {table} ADD FOREIGN KEY  ({col_name_fk}) REFERENCES {table_pk}({col_name_pk});')    

define_FK('orpha_definitions','OrphaCode','orphanet.orpha_diseases','OrphaCode')
```

```python
#define foreign key in clinical annotations table
define_FK('clinical_annotations','OrphaCode','orphanet.orpha_diseases','OrphaCode')
define_FK('clinical_annotations','HPO_ID','orphanet.HPO_ontology','ID')
```
